```python
## get the urls link range 1-50 pages

base_url='https://www.thestar.com.my/search/?q=HIV&qsort=oldest&qrec=10&qstockcode=&pgno='
url_list=["{}{}".format(base_url, str(page)) for page in range(1,50)]

s=[]
for Url in url_list:
#     print(Url)
    s.append(Url)
s
```

### json.dump() ,json.dumps()
- The **json.dump(data,file)** function takes in two arguments:

- **Data** that needs to be written to a JSON file.
- A **file** object that can be used to save the data

- In this program, we store the set of numbers in urljson.json. The extension .json shows that the file contains data in JSON format.

- We then access the file in 'w' mode (write mode), to make enable data to be written into a JSON file. Finally, the json.dump() function stores the set of numbers in the file numbers.json file.

- The **dump()** method takes two parameters (data and file), while the **dumps()** method takes only one parameter (data)
- The dump() method is combined with file operations, unlike the dumps() method.

### json.load(), json.loads()

- We use the json.load function to read a JSON file.

- The **json.load()** function takes one argument which is the file object.
json.load(file_object)

- The json.load() function parses the JSON file and returns a dictionary named data.

```python
import json
## write json file

file='urljson.json'   #use the file extension .json
with open(file,'w') as file_object: #open the file in write mode
    json.dump(s,file_object)  # json.dump() function to stores the set of numbers in numbers.json file
```

```python
art_name=[]
art_date=[]
for pg in s:
    
    #query the website and return the html to the variable 'page'
    page = urllib.request.urlopen(pg)
    
   # parse the html using beautiful soap and store in variable `soup`
    soup = BeautifulSoup(page,'html.parser')
    
    
    
    name = [x.get_text(strip=True) for x in soup.find_all("h2", {"class": "f18"})]
    dates= [x.get_text(strip=True) for x in soup.find_all("span", {"class": "timestamp"})]
    #save the data in tuple
    art_name.append(name)
    art_date.append(dates)
```

```python
article_title=pd.DataFrame(art_name,columns=['page1','page2','page3','page4','page5','page6','page7','page8','page9','page10'])
article_title
```

```python
article_Date=pd.DataFrame(art_date,columns=['page1','page2','page3','page4','page5','page6','page7','page8','page9','page10'])
article_Date
```
