> ## Feature Engineering
> ### select needy features

```python
df=pd.DataFrame(data,columns=['reviewText','summary','overall'])
df.head()
```

```python
## combine reviewtext and summary cols

df['text']=df['reviewText']+' '+df['summary']
df['text'][0]
```

```python
df['rating']=df['overall']

del df['reviewText']
del df['summary']
del df['overall']

df.head()
```

```python
df.rating.value_counts()
```

> ## sentiment analysis
- convert ratings 3,4,5 into 1 as positive reviews and 2,1 into 0 as negative reviews

```python
def ratings(rating):
    if rating<3:
        return 0
    else:
        return 1
df['ratings']=df['rating'].apply(ratings)
df=df.drop(['rating'],axis=1)
df.head()
```

> Cleaning the text
- remove stopwords
- tokenization
- lemmatize
- lowercase
- tfidf

```python
lemma=WordNetLemmatizer()
stop=set(stopwords.words('english'))
punctuation=list(string.punctuation)
stop.update(punctuation)


def lema_fun(text):
    new_text=[]
    for word in text.split():
        if word .lower() not in stop:
            lem=lemma.lemmatize(word)
            new_text.append(lem.lower())
    return ' '.join(new_text)

df.text=df.text.apply(lema_fun)
```

> #### Handle imbalanced dataset


```python

#Balancing data using SMOTE ..oversampling

from imblearn.over_sampling import SMOTE

smote=SMOTE(random_state=42,sampling_strategy=1)
x_bl,y_bl=smote.fit_resample(x_tf,Y)

print(y_bl.value_counts())

print(f'Trianing target: {Counter(y_bl)}')
print(f'Testing target : {Counter(Y)}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from collections import Counter
```

> ### Model Trianing

```python
x_train,x_test,y_train,y_test=train_test_split(x_bl,y_bl,test_size=0.3,random_state=0)

print(f'Trianing target: {Counter(y_train)}')
print(f'Testing target : {Counter(y_test)}')
```

```python
#Logistic Regression

log=LogisticRegression()
log_mod=log.fit(x_train,y_train)
log_pred=log.predict(x_test)

log_acc=accuracy_score(y_test,log_pred)


print('logistic test accuracy: ',log_acc)
```

```python
print('classification report:\n',classification_report(y_test,log_pred))
```

```python
cm=confusion_matrix(y_test,log_pred)
cm
```

```python
# Naive Bayes

nb = MultinomialNB()
nb_model = nb.fit(x_train, y_train)
nb_predict = nb.predict(x_test)


nb_acc = accuracy_score(y_test,nb_predict)

print('nb test accuracy:', nb_acc)
```

```python
print('classification report:\n',classification_report(y_test,nb_predict))
```

```python
cm=confusion_matrix(y_test,nb_predict)
cm
```
